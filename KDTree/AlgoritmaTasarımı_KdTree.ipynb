### Kd Tree Nearest Neighbor Search Algorithm 

2 boyutlu (2d) en yakın komşu bulma algoritması:

```python
from pyflann import *
import matplotlib.pyplot as plt
```

```python

dataset = np.random.rand(100,2).astype(np.float) *10 # 100 adet 2 boyutlu nokta olustur

testset = np.array([[5,5]]).astype(np.float)
```

```python
flann = FLANN() # flann instance'i olustur
result, dists = flann.nn(dataset, testset, 10, algorithm="kmeans", branching=32, iterations=7, checks=16) 
    
```

```python
# komsu ara
print(result) # sonuc indisleri
print(dists) # her indise olan uzaklik
```

```python
result_set = dataset[result[0]] # sadece ilk nokta icin gorsellestirme

# Bu kodda zaten bir nokta oldugu icin bir sey fark etmeyecek.

# Gorsellestirme
plt.scatter(dataset[:,0],dataset[:,1],c='red',lw=0,s=100.0)
plt.scatter(result_set[:,0],result_set[:,1],c='green',lw=0,s=100.0)
plt.scatter(testset[0][0],testset[0][1],c='blue',lw=0,s=100.0)
plt.legend(["Veri Kumesi", "Yakin Noktalar","Hedef Nokta"])
plt.show()
```
